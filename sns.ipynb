```python
import seaborn as sns
import pandas as pd
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
planets = sns.load_dataset("planets")
titanic = sns.load_dataset("titanic")
```

```python
tips
```

```python
sns.scatterplot(x="day", y="total_bill", data=tips, hue="day" , size = "size", palette="YlGnBu")
# x and y are used as axis , data for loading the data set
#hue to represent days in diff colors
#size represents the size of diff data points acc to size column
#pallete to diff with colors
```

```python
sns.histplot(tips['tip'] ,  kde=True, bins=15)
#kde represents the line going through the bars
#bins represent the no. of bars
```

```python
sns.distplot(tips['tip'] ,  kde=True, bins=15)
```

```python
sns.displot(tips['tip'] ,  kde=True, bins=15)
#displot is like histplot but bigger in size
```

```python
sns.barplot(x="sex" ,  y="tip" , data=tips, hue="sex",  palette="YlGnBu")
```

```python
sns.boxplot(x="day" , y="tip" , data=tips , hue="sex" , palette="YlGnBu")
```

```python
sns.stripplot(x="day" , y="tip" , data=tips , hue="sex" , palette="YlGnBu" , dodge=True)
#dodge differentitate the male and female points
```

JOIN PLOTS

```python
sns.jointplot(x="tip" , y="total_bill" , data=tips)
```

```python
sns.jointplot(x="tip" , y="total_bill" , data=tips , kind="reg")
#kind="reg" shows a regression line
```

```python
sns.jointplot(x="tip" , y="total_bill" , data=tips , kind="kde" , shade=True ,  fill=True ,  cmap= "YlGnBu")
#kind="reg" shows a regression line
#kind="kde" gives a 3dlook
#shade and fill used to give shades to the 3d look
#cmap is like pallete
```

```python
sns.jointplot(x="tip" , y="total_bill" , data=tips , kind="hex" ,  cmap= "YlGnBu")
#hex shapes 
```
